```python
from fbprophet import Prophet
```

```python
data.columns
```

```python
total_cases=data.groupby('date').sum()['total_cases'].reset_index()
total_deaths=data.groupby('date').sum()['total_deaths'].reset_index()
hosp_patients=data.groupby('date').sum()['hosp_patients'].reset_index()
```

```python
total_cases.columns=['ds','y']
total_cases['ds']=pd.to_datetime(total_cases['ds'])
```

```python
total_cases.tail()
```

```python
m=Prophet(interval_width=0.95)
m.fit(total_cases)
future=m.make_future_dataframe(periods=7)
future.tail()
```

```python
forecast=m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
total_cases_forecast_plot=m.plot(forecast)
```

```python
total_cases_forecast_plot=m.plot_components(forecast)
```
